```python
import requests  
from bs4 import BeautifulSoup
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

def getdata(url):  
    r = requests.get(url)  
    return r.text  

# Read in DF and create two new dataframed one holding the card names and one holding the which set each card belongs to
path = "USETHIS_Outside_folder.csv"
df = pd.read_csv(path, encoding="utf-8")
df_name = df[["Name"]]
df_set = df["Set"].tolist()

# Fill mtg_cards with names and sets
mtg_cards = {}
for card in df_name["Name"]:
    mtg_cards[card] = []
count = 0
for card in mtg_cards:
    mtg_cards[card].append({'Set' : df_set[count]})
    count += 1

# Web Scraping    
for card in mtg_cards:
    try:
        # Scrapes image url of each card from scryfall
        driver = webdriver.Chrome()
        driver.get("https://scryfall.com/advanced")
        search_box = driver.find_element_by_name("name")
        search_box.send_keys(str(card))
        search_box = driver.find_element_by_name("name")
        set_box = driver.find_element_by_xpath("/html/body/div[3]/form/div/div[9]/div/div[1]/span/span[1]/span/ul/li/input")
        set_box.send_keys(str(mtg_cards[card][0]['Set']))
        set_box.send_keys(Keys.DOWN)
        set_box.send_keys(u'\ue007')
        search_button = driver.find_element_by_xpath("//button[@class='button-n submit-n']")
        search_button.click()
        url = driver.current_url
        htmldata = getdata(url)  
        soup = BeautifulSoup(htmldata, 'html.parser') 
        for item in soup.find_all('img'): 
            image_url = item['src']
            mtg_cards[card].append({'Image URL' : image_url})
            mtg_cards[card][1]['Image URL']
        driver.quit()
    #     This will output clickable image url
    #     print(cards[card][1]['Image URL'])
    # If a card is not found print ERROR msg and save card name to list of mtg cards not found
    except:
        print("ERROR")
```

```python
# Reformats the names to a format that is easy to put into a dataframe
card_name_col = []
for card in mtg_cards:
    card_name_col.append({'Name' : card})
card_name_col

card_set_col = []
for card_set in mtg_cards:
    card_set_col.append(mtg_cards[card_set][0])
card_set_col

card_image_url_col = []
for card_image_url in mtg_cards:
    card_image_url_col.append(mtg_cards[card_image_url][1])

# Create DFs
card_name_df = pd.DataFrame(card_name_col)
card_name_df

card_set_df = pd.DataFrame(card_set_col)
card_set_df

card_image_url_df = pd.DataFrame(card_image_url_col)
card_image_url_df

card_object_data = pd.concat([card_name_df,card_set_df,card_image_url_df], axis=1)
card_object_data

# Export the cleaned data to a csv
card_object_data.to_csv("card_object_data.csv", index=False, header=True)
```
